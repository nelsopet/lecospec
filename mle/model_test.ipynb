```r
speclib_path <- "./mle/trainingData.csv"
training_data <- read.csv(speclib_path) %>%
    dplyr::select(Functional_group2, everything()) %>% 
    dplyr::select(-ScanID:-Functional_group2_Freq) %>% 
    rename(Classes = Functional_group2) %>%
    mutate(Classes = as.factor(Classes)) %>% 
    as.data.frame()

summary(training_data)
```

```r
print(colnames(training_data))
write.csv(colnames(training_data), "./mle/colnames.csv")
```

```r
rf_FncGrp2 <- ranger::ranger(
    Classes ~ .,
    data = training_data,
    num.trees = 1000
    )
```

```r
save_path <- "./noisyModel.rds"
save(rf_FncGrp2, save_path)
```
