```r
setwd("~/GitHub/lecospec/")
source("./Functions/lecospectR.R")
```

```r
training_path <- "Data/v2/train_5nm_125.csv"
test_path <- "Data/v2/test_5nm_125.csv"

test_data_full <- read.csv(test_path)
test_labels <- test_data_full$FncGrp1 %>% as.factor()
test_data <- subset(
    test_data_full,
    select = -c(
        X,
        UID,
    	FncGrp1,
        Site
        ))


train_data_full <- read.csv(training_path)
train_data <- subset(
    train_data_full,
    select = -c(
        X,
        UID,
    	FncGrp1,
        Site
        ))
labels <- train_data_full$FncGrp1 %>% as.factor()
```

```r
variable_importance_model <- ranger::ranger(
        num.trees = 1000,
        importance = "impurity_corrected",
        replace = TRUE,
        classification = TRUE,
        x = train_data,
        y = labels
    )
```

```r
print(names(variable_importance_model))
```

```r
variable_importance <- sort(variable_importance_model$variable.importance, decreasing = TRUE) %>% as.data.frame()
variable_importance
```

```r
print(
    xtable::xtable(
        variable_importance,
    ), 
        type="html", 
        file="./temp_variable_importance_table.html")
```
