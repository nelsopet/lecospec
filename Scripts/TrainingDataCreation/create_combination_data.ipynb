```r
get_bands_only_version <- function(filepath, save_path){

    data <- read.csv(filepath, header = TRUE)

    variables <- colnames(data)

  
    band_names <- c(
            "NDVI","OSAVI","SAVI","MTVI","NDWI","PWI",
            "MSI", "SRWI","GMI1","GMI2","MCARI","TVI",
            "Vogelmann4","Boochs","Boochs2",
            "D1", "D2", "Datt5", "Datt6", "DD",
            "DDn", "EGFN", "EGFR", "EVI", "GI",
            "Gitelson", "Gitelson2", "GreenNDVI", "MCARIOSAVI",
            "CARI","CI","Carter","Carter2","Carter3","Carter4",
            "Carter5","Carter6","Datt","Datt2","Datt3","Datt4",
            "MCARI2","MCARI2OSAVI2","mNDVI","mND705","Maccioni",
            "mREIP","MSAVI","mSR","mSR705","mSR2","MTCI","NDVI2",
            "NDVI3","NPCI","OSAVI2","RDVI","REPLE","REPLi",
            "SIPI","SPVI","SR","SR1","SR2","SR3","SR4","SR5","SR6",
            "SR7", "SR8","SRPI","SumDr1","SumDr2","TCARI","TCARI2",
            "TCARIOSAVI","TCARI2OSAVI2","Vogelmann","NDLI",
            "Vogelmann2","Vogelmann3","PRI","CAI","NDNI",
            "PSSR", "PSND", "CRI1", "CRI2", "CRI3",
            "CRI4", "MPRI", "PRICI2", "CI2", "PSRI", "ClAInt", 
            "TGI", "PRInorm","PARS","DPI","Datt7","Datt8",
            "GDVI2","GDVI3","GDVI4","LWVI1","LWVI2",
            "DWSI1","DWSI2","DWSI3","DWSI4","DWSI5",
            "SWIRFI", "SWIRLI", "SWIRSI", "SWIRVI"
        )
    
    new_variables <- setdiff(
        variables,
        band_names
    )
    filtered_data <- data[,new_variables]
    write.csv(filtered_data, file=save_path, row.names=FALSE)

    return(filtered_data)

}
```

```r
input_path <- "Data/gs/x_train/grd_raw_corrected.csv"
output_path <- "Data/gs/x_train/grd_corrected_only_bands.csv"

filtered_data <- get_bands_only_version(input_path, output_path)
print(colnames(filtered_data))
```
